```python
with open('eff_squre52_data.json') as f:
    efficiency = json.load(f)
```

```python
fig, ax = plt.subplots(1,2, figsize=(9,3.5), constrained_layout=True)
xaxis='energy, keV'
yaxis=['Absorption efficiency, %', 'Photo-peak efficiency, %']
en2=[20, 50, 100, 200, 500, 1000, 4000]
ticks=[[20, 30, 40,50,60,70,80,90,100],[10, 20, 30,40,60,80,100]]

for a in efficiency:
    ax[0].plot(efficiency[a]['en'], efficiency[a]['abs_eff'], linewidth=2, label=efficiency[a]['size'])
    ax[1].plot(efficiency[a]['en'], efficiency[a]['ph_eff'], linewidth=2, label=efficiency[a]['size'])
    
ax[0].set(ylabel=yaxis[0],yscale='log')
ax[1].set(ylabel=yaxis[1],yscale='log')

for i in ax:
    i.grid(True)
    i.set(xlabel=xaxis, xscale='log')
    i.legend()
    i.get_yaxis().set_major_formatter(tick.ScalarFormatter())

    i.get_xaxis().set_major_formatter(tick.ScalarFormatter())
    i.set_xticks(en2)
ax[0].set_yticks(ticks[0])
ax[1].set_yticks(ticks[1])

plt.show()
#plt.savefig('abs_photo_eff_sq51.pdf', format='pdf')
```
